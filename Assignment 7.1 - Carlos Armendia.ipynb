```python
import numpy as np
import pandas as pd
from __future__ import print_function

df = pd.read_excel(
    "https://docs.google.com/spreadsheets/d/1OGs8KoHGEZDxyFdaxAKmZbeO1BiSWVQ6/export?format=xlsx", sheet_name = "Canada by Citizenship", 
    skiprows = range(20), skipfooter = 2)

def display_no_index(df):
    return df.style.hide(axis='index')
    
df.head()
```

```python
df = df.drop(columns=['Type', 'Coverage', 'AREA', 'REG', 'DEV'])
position = 0
empty_row = pd.DataFrame([[''] * len(df.columns)], columns=df.columns)
df = pd.concat([df.iloc[:position], empty_row, df.iloc[position:]], ignore_index=True)

df.columns.values[0] = ''
df.columns.values[1] = 'Continent'
df.columns.values[2] = 'Region'

df.iloc[0, 0] = 'Country'

df['TOTAL'] = df.iloc[:, 4:].sum(axis=1)

display_no_index(df.head())
```
